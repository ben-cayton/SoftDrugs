```python
import pandas as pd
import numpy as np

def load_data(csv_file='./data.csv'):
    loaded_data = pd.read_csv(csv_file)
    n_files = len(loaded_data)
    Ids = loaded_data["Ids"]
    labels = loaded_data["Labels"]
    labels = labels.to_numpy()
    return Ids, labels
```

```python
import deepchem as dc
import os
from os.path import dirname, join as pjoin
from deepchem.feat.complex_featurizers.contact_fingerprints import ContactCircularFingerprint

def featurizing_ccfp(Id_array, label_array, data_folder_name):
    n_samples = len(Id_array)
    features = np.zeros(n_samples)
    labels = np.zeros(n_samples)
    prim_path = ('C:\\Users\\joslynn.deaver\\SoftDrugs\\Sanitization')
    data_dir = pjoin(prim_path, data_folder_name)
    os.chdir(data_dir)
    ccfp_featurizer = dc.feat.ContactCircularFingerprint(size=2048)
    for idx in range(n_samples):
        pdbid = Id_array[idx]
        protein = '%s.pdb' % (pdbid)
        ligand = '%s_ligand.pdb' % (pdbid)
        try:
            print(protein, ligand)
            features[idx] = ccfp_featurizer.featurize(zip(ligand, protein)).flatten()
            labels[idx] = label_array[idx]
        except:
            print('%s Featurization Failed' % (pdbid))
    os.chdir(prim_path)
    features_and_labels = {'Feature Array': features, 'Label': labels}
    features_and_labels = pd.DataFrame(data=features_and_labels)
    features_and_labels.to_csv('featured_ccfp_train_data.csv')
```
