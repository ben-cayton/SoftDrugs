```python
import pandas as pd
import numpy as np

def load_data(csv_file='./data.csv'):
    loaded_data = pd.read_csv(csv_file)
    n_files = len(loaded_data)
    Ids = loaded_data["Ids"]
    labels = loaded_data["Labels"]
    labels = labels.to_numpy()
    return Ids, labels
```
